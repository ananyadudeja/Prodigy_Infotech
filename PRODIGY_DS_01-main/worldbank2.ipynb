```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'C:\prodigy_infotech_01\API_SP.POP.TOTL_DS2_en_csv_v2_5871594.csv')
```

```python
df
```

```python

# Count the occurrences of each unique value in the first column
first_column_counts = df.iloc[:, 0].value_counts()

# Plotting
bar_width = 0.9
x = range(len(first_column_counts.index))

plt.bar(first_column_counts.index, first_column_counts.values, width=bar_width)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')
plt.xticks(x, first_column_counts.index, rotation=45)
plt.tight_layout()
plt.show()
```
